```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv('baseline2.gz')
data['type'].unique()
```

```python
personEntersVehicle = data[(data['type'] == 'PersonEntersVehicle') & (data['time'] <= 86400)].dropna(how='all', axis=1)
```

```python
personEntersVehicle['lastSeven'] = personEntersVehicle['vehicle'].str[-7:]
```

```python
personEntersVehicle['firstTwo'] = personEntersVehicle['vehicle'].str[:2]
```

```python
personEntersVehiclePassenger = personEntersVehicle[personEntersVehicle['person'].str.contains('Driver') != True]
```

```python
personEntersVehicleVT = personEntersVehicle[personEntersVehicle['firstTwo'] == 'VT']
```

```python
personEntersVehicleVTDriver = personEntersVehicleVT[personEntersVehicleVT['person'].str.contains('Driver') == True]
```

```python
personEntersVehicleVTPassenger = personEntersVehicleVT[personEntersVehicleVT['person'].str.contains('Driver') != True]
```

```python
personEntersVehicleVTPassenger
```

```python
sumUp = personEntersVehiclePassenger.groupby('firstTwo').count().reset_index()
```
