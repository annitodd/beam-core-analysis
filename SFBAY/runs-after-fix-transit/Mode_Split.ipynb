```python
import pandas as pd
```

```python
dataDict = {}

for i in range(0, 13):
    data = pd.read_csv('1.realizedMode (' + str(i) + ').csv')
    data = data[data['hours'] <= 24]
    data.loc['sum'] = data.sum(axis=0)
    data['sum'] = data.sum(axis=1)
    data.loc['rate'] = data.loc['sum'] / data['sum']['sum']
    data = data.drop(['hours'], axis=1)
    dataDict['1.realizedMode (' + str(i) + ').csv'] = data.loc['rate']
pd.concat(dataDict).to_csv('out/modesplit.csv')
```
