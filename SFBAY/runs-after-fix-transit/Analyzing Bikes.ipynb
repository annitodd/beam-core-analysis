```python
import pandas as pd
```

```python
data = pd.read_csv('0.3pop-sfbay-fewer-bikes.gz.gz')
replanning = data[data['type'] == 'Replanning'].dropna(how='all', axis=1)
bikeReplanning = replanning[replanning['reason'].str.contains('BIKE')]
bikeReplanning = bikeReplanning[bikeReplanning['reason'].str.contains('ResourceUnavailable BIKE')]
bikeReplanning = bikeReplanning[bikeReplanning['reason'].str.contains('TRANSIT') == False]
display(bikeReplanning)
```
