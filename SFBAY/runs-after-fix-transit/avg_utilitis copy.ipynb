```python
from pandas import DataFrame
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
utilitis = {}
for i in range(0, 13):
    count = []
    data = pd.read_csv('1.modeChoiceAlternativesWhenRHPooled (' + str(i) + ').csv.gz')
    deniedUtilities = data[(data['transit'].str.startswith('NaN') == False) & (data['wasChosen'] == 0)]['altUtility'].mean()
    count.append(deniedUtilities)
    realizedUtilities = data[(data['transit'].str.startswith('NaN') == False) & (data['wasChosen'] == 1)]['altUtility'].mean()
    count.append(realizedUtilities)
    utilitis[i] = count
    print(utilitis)


d = pd.DataFrame(utilitis) 
d.to_csv('out/Transit-Trips.csv',index=False,mode='a',header=None)
```
